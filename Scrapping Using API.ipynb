## Scapping using API

```python
with open("API.txt", "r") as file:
    api_key = file.readline().strip()  # Read the first line and remove any whitespace/newline

api_key
```

```python

API_KEY = api_key
HEADERS = {
    "x-api-key": API_KEY,
    "Accept": "application/json"
}

ARTIST_MBID = "27e2997f-f7a1-4353-bcc4-57b9274fa9a4"  # Babymetal's MusicBrainz ID
BASE_URL = f"https://api.setlist.fm/rest/1.0/artist/{ARTIST_MBID}/setlists"

def get_setlists(page=1):
    params = {"p": page}
    response = requests.get(BASE_URL, headers=HEADERS, params=params)

    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None

# Example: Get first 3 pages of setlists
all_setlists = []
for page in range(1, 49):
    data = get_setlists(page)
    if data and 'setlist' in data:
        all_setlists.extend(data['setlist'])

print(f"Retrieved {len(all_setlists)} setlists")
```

```python
import json

# Save to a JSON file
with open("babymetal_setlists.json", "w", encoding="utf-8") as f:
    json.dump(all_setlists, f, ensure_ascii=False, indent=2)

print("💾 Setlists saved to 'babymetal_setlists.json'")
```

```python
import json
import pandas as pd

# Load the JSON
with open("babymetal_setlists.json") as f:
    data = json.load(f)

# Normalize JSON to get flat structure
df_info = pd.json_normalize(data)

df_info
```

```python
# Normalize JSON to get flat structure
# make this in one list, jadi nanti pas vectorize bisa 
import numpy as np

list_song = []
for x in range(df_info.shape[0]):
    try:
        # Try to extract song names from the nested JSON structure
        songs = df_info.iloc[x]['sets.set'][0]['song']
        song_names = list(pd.json_normalize(songs)['name'])
        list_song.append(song_names)
    except (IndexError, KeyError, TypeError):
        # If any error occurs (e.g., missing data), append NaN
        list_song.append(np.nan)
list_song
```

## Create Target from Song

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
dummies = pd.DataFrame(mlb.fit_transform(df_info['song'][2:]), 
                      columns=mlb.classes_,
                      index=df_info[2:].index)
dummies
```

```python
df = df_info[2:].copy()
df = pd.concat([df,dummies],axis=1)
df
```
